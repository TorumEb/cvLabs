```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
def plot_image(img):
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

    plt.imshow(img)
    plt.axis('off')
    plt.show()
```

```python
left_img = cv.imread('img0.jpg', cv.IMREAD_COLOR )
right_img = cv.imread('img1.jpg', cv.IMREAD_COLOR )
```

```python
templ_size = 10
templ = left_img[:,-templ_size: ,:]
res = cv.matchTemplate(right_img,templ,cv.TM_CCOEFF)

min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
print(max_loc)
result_img = np.zeros((left_img.shape[0],left_img.shape[1]+right_img.shape[1] - max_loc[0] -templ_size, left_img.shape[2] ),dtype=np.uint8)
result_img[:,0:left_img.shape[1],:] = left_img
result_img[:,left_img.shape[1]:,:] = right_img[:,max_loc[0]+templ_size:,:]
plot_image(result_img)
print(result_img.shape[0], result_img.shape[1])
```
